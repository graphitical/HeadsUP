# PDF Title Converter
This script reads in a PDF with a name we don't like (e.g. 12386745.23948723.pdf), reads its title, and then resaves it as a new PDF with the title as the file name.

```python
import PyPDF2 as ppdf
from pathlib import Path
import os
```

```python
# Change to whatever paths you care about
BASE_PATH = '/Users/danielrevier/Downloads'
DIRS = ['2019']

# make new dirs for saving files
os.chdir(BASE_PATH)
[os.mkdir(os.path.join(BASE_PATH, DIR)+'_retitled') for DIR in DIRS if not os.path.exists(os.path.join(BASE_PATH, DIR)+'_retitled')];
```

```python
n = 0
for DIR in DIRS:
    for file in Path(DIR).rglob('*.pdf'):
        print('Old Name:',file.name)
        input_pdf = ppdf.PdfFileReader(os.path.join(DIR,file.name),'rb')
        
        pdf_title = input_pdf.getDocumentInfo().title
        print('PDF Title:',pdf_title)
        # manipulate title to be whatever you want
        pdf_title = ''.join(filter(str.isalnum, pdf_title))

        # new path to save retitled pdf file
        output_path = os.path.join(BASE_PATH,DIR+'_retitled',pdf_title+'.pdf')
        print('Out Path:',output_path)
        
        # copy current pdf
        output_pdf = ppdf.PdfFileWriter()
        for page in input_pdf.pages:
            output_pdf.addPage(page)
            
        # save pdf
        with open(output_path,'wb') as output_file:
            output_pdf.write(output_file)
            
       # debugging
#         if n > 2: break
        
#         n += 1
```
